```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from statsmodels.graphics import tsaplots
from tg.datasets import DATASET_FACTORY_LOOKUP
from tg import get_root_path
from tg.datasets import DATASET_FACTORY_LOOKUP

color_palette = list(sns.color_palette("husl", 9).as_hex())
```

```python
t = np.linspace(0, 100, num=50)

function_name = 'tendencia_linear'
base_function = 0.5*t
function_trend = base_function + np.random.normal(0, 2, size=t.shape)

fig, ax = plt.subplots(figsize=(10, 6))

plt.plot(t, base_function, linestyle='--', color=color_palette[0], linewidth=1.5)
plt.plot(t, function_trend, 'o', linestyle='-', linewidth=1.5, markersize = 3.0, color="black",  alpha=0.6)
plt.tight_layout()

plt.ylabel('Série Temporal', fontsize=12)
plt.xlabel('Tempo', fontsize=12)

ax.xaxis.set_ticklabels([])
ax.yaxis.set_ticklabels([])
# ax.grid(linewidth=0.10)

plt.savefig(get_root_path("reports/{}.png".format(function_name)), dpi=300, bbox_inches='tight')

plt.show()
```

```python
t = np.linspace(1, 100, num=70)

function_name = 'tendencia_nula'
base_function = np.array([0.5] * len(t))
function_trend = base_function + np.random.normal(0, 0.1, size=t.shape)

fig, ax = plt.subplots(figsize=(10, 6))

plt.plot(t, base_function, linestyle='--', color=color_palette[0], linewidth=1.5)
plt.plot(t, function_trend, 'o', linestyle='-', linewidth=1.5, markersize = 3.0, color="black",  alpha=0.6)
plt.tight_layout()

plt.ylabel('Série Temporal', fontsize=12)
plt.xlabel('Tempo', fontsize=12)

ax.xaxis.set_ticklabels([])
ax.yaxis.set_ticklabels([])
# ax.grid(linewidth=0.10)

ax.set_ylim([0.1, 0.9])

plt.savefig(get_root_path("reports/{}.png".format(function_name)), dpi=300, bbox_inches='tight')

plt.show()
```

```python
series = DATASET_FACTORY_LOOKUP["AIR_PASSENGERS"]()
acf = tsaplots.acf(series, nlags = 100)

fig, axs = plt.subplots(2, 1, figsize=(10, 6))

axs[0].plot(series, linestyle='-', linewidth=1.5, markersize = 3.0, color="blue",  alpha=1.0)
axs[1].stem(acf, 'blue', markerfmt=' ')

axs[0].set_ylabel('Número de passageiros', fontsize=12)
axs[0].set_xlabel('Data', fontsize=12)

axs[1].set_ylabel('ACF', fontsize=12)
axs[1].set_xlabel('Índice', fontsize=12)

axs[0].grid(linewidth=0.10)
axs[1].grid(linewidth=0.10)

plt.tight_layout()
plt.savefig(get_root_path("reports/ts_air_passengers.png"), dpi=300, bbox_inches='tight')

plt.show()
```
