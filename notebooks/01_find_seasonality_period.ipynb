```python
import numpy as np
from statsmodels.graphics import tsaplots
from scipy.signal import argrelextrema
from tg.datasets import DATASET_FACTORY_LOOKUP
```

# Finding seasonality period

```python
for dataset_name, factory in DATASET_FACTORY_LOOKUP.items():
    ts = factory()
    tsaplots.plot_acf(ts, lags = 50);
    acf = tsaplots.acf(ts, nlags = 100)
    local_max_idxs = argrelextrema(acf, np.greater)[0]
    print(f"{dataset_name}: {local_max_idxs}")
```
