```python
import glob

import matplotlib.pyplot as plt
import mlflow
import numpy as np
import pandas as pd
import seaborn as sns

from tg import get_data_path, get_root_path
from tg.datasets import DATASET_FACTORY_LOOKUP

# color_palette = list(sns.color_palette("pastel").as_hex())
color_palette = list(sns.color_palette("husl", 9).as_hex())

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
t = np.linspace(0, 100, num=50)

function_name = 'linear_trend'
base_function = 0.5*t
function_trend = base_function + np.random.normal(0, 2, size=t.shape)

fig, ax = plt.subplots(figsize=(10, 6))

plt.plot(t, base_function, linestyle='--', color=color_palette[0], linewidth=1.5)
plt.plot(t, function_trend, 'o', linestyle='-', linewidth=1, markersize = 2.5, color="black",  alpha=0.6)
plt.tight_layout()

plt.ylabel('Série Temporal', fontsize=12)
plt.xlabel('Tempo', fontsize=12)

ax.xaxis.set_ticklabels([])
ax.yaxis.set_ticklabels([])
# ax.grid(linewidth=0.10)

plt.savefig(get_root_path("reports/{}.png".format(function_name)), dpi=300, bbox_inches='tight')

plt.show()
```
