```python
import numpy as np
import pandas as pd
```

```python
t = np.linspace(0, 150, num=150)
s = pd.Series(np.sin(t * 2 * np.pi / 30))
s.index.name = "index"
s.name = "sine"
s.plot()
```

```python
t = np.linspace(0, 150, num=150)
s = pd.Series(np.sin(t * 2 * np.pi / 30) + np.random.randn(t.size) / 5)
s.index.name = "index"
s.name = "sine"
s.plot()
```
